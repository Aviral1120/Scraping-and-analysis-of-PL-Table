```python
#positioning of teams on the basis of Goal Difference(F-A)
#to avoid the permanent sorting of the whole dataframe we use it's copy to carry out the operation
s1 = df.copy()
GD = s1.sort_values(by='GD', ascending=False)
GD.head(20)
```

```python
#positioning of teams on the basis of Games Played
s2 = df.copy()
GP = s2.sort_values(by='GP', ascending=False)
GP.head(20)
```

```python
#positioning of teams on the basis of wins
s3 = df.copy()
W = s3.sort_values(by='W', ascending=False)
W.head(20)
#similary we can do it other columns too
```
